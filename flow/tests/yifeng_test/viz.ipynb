```python
import pandas as pd
import matplotlib.pyplot as plt

df_analyse = pd.read_csv('output_freenode.csv')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# extra_speed = 15 # [10,15,20,25]
for extra_speed in [10,15,20,25]:
    df_plot = df_analyse[df_analyse['speed_limit_extra'] == extra_speed] [df_analyse['flow_rate'] <=400] # [df_analyse['flow_rate'] >=1000]
    df_plot
    # 
    plt.figure(figsize=(12, 6))

    #  Average Outflows vs Flow Rate
    plt.subplot(1, 2, 1)  # 121
    #  'allin'
    sns.lineplot(data=df_plot[df_plot['road_mode'] != 'allin'], x='flow_rate', y='avg_outflows', hue='road_mode', style=True, dashes=True)
    #  'allin' 
    sns.lineplot(data=df_plot[df_plot['road_mode'] == 'allin'], x='flow_rate', y='avg_outflows', color='red', linewidth=2.5, label='allin')
    plt.title(f'Average Outflows with Extra Speed Limit {extra_speed}')

    #  Average Velocity vs Flow Rate
    plt.subplot(1, 2, 2)  # 122
    #  'allin'
    sns.lineplot(data=df_plot[df_plot['road_mode'] != 'allin'], x='flow_rate', y='avg_velocity', hue='road_mode', style=True, dashes=True)
    #  'allin' 
    sns.lineplot(data=df_plot[df_plot['road_mode'] == 'allin'], x='flow_rate', y='avg_velocity', color='red', linewidth=2.5, label='allin')
    plt.title('Average Velocity with Extra Speed Limit {}'.format(extra_speed))

    # 
    plt.show()
```

```python
# flow_rate as x-axis, avg_outflows as y-axis, color by road_mode bar plot
plt.figure(figsize=(12, 6))
sns.barplot(data=df_plot, x='flow_rate', y='avg_outflows', hue='road_mode')
plt.title('Average Outflows vs Flow Rate')
plt.show()
```

```python
# path_choose 
for i in df_plot.index:
    path = df_plot.loc[i,'link']
    flow_rate = df_plot.loc[i,'flow_rate']
    # df_raw = pd.read_csv('./data/New Folder With Items 2/' + path[7:])
    df_raw = pd.read_csv('./data/' + path[7:])
    mode = df_plot.loc[i,'road_mode']
    df_raw_ana = df_raw.groupby(['id','edge_id']).size()
    df_raw_ana = df_raw_ana.reset_index()
    df_raw_ana.columns = ['id','edge_id','count']
    df_raw_plot = df_raw_ana.groupby('edge_id').size().reset_index()
    df_raw_plot.columns = ['edge_id','count']
    # print(df_raw_plot)
    # plot df_raw_plot with bar_chart with value of ['left_vertical','center_vertical','center_up']
    plt.figure(figsize=(12, 6))
    sns.barplot(data=df_raw_plot[df_raw_plot['edge_id'].isin(['left_vertical','center_vertical','center_up'])], x='edge_id', y='count')
    plt.title(f'flow rate in {flow_rate},with mode as {mode}')
    plt.show()

    break
```
